<h3>Немного о строках</h3>

Строки в Python — это неизменяемые объекты, которые состоят из последовательности символов. Строки заключаются в кавычки — одинарные (‘’) или двойные (“”).
Каждый символ строки имеет свой индекс, для первого символа он нулевой.

```python
s = 'abcdef' # это строка
s = "abcdef" # и это такая же строка
print(s[3]) # а это операция обращения к элементу строки
```

<h3>Индексы</h3>

```python
a = [1, 2, 3, 4, 5] # например в массиве можно просто поменять элемент
a[0] = 0
print(a)
```

<h3>Перевод в двоичную СС</h3>

Перевод числа в двоичную систему счисления осуществляется с помощью функции bin, которая принимает целое десятичное число, и возвращает строку, являющуюся двоичной записью исходного числа. 

```python
for n in range(5, 8):
    print(bin(n))
```

В примере выше перед числом присутствует префикс — два символа '0b', которые обозначают двоичную СС. Префикс добавляется по умолчанию и его необходимо убрать. Самый простой способ сделать это — срез:

```python
for n in range(5, 8):
    s = bin(n)[2:]
    print(s)
```

<h3>Сумма цифр двоичной строки</h3>

В задачах часто фигурирует такая фраза: <i>"складываются все цифры двоичной записи"</i>. Далее могут попросить найти остаток от деления этой суммы на 2.

<b>В двоичной системе счисления есть только две цифры: 0 и 1. Поэтому при сложении всех цифр числа сумма будет равна количеству единиц в двоичной записи числа.</b>

Метод count возвращает количество вхождений подстроки в строку. Именно количество единиц в двоичной записи числа и будет суммой его цифр.


<h3>Перевод в десятичную СС</h3>

Так как результатом работы алгоритма является десятичное число R, то после всех манипуляций с двоичной строкой, необходимо вернуться обратно в 10-ую СС. Делается это с помощью функции int()

```python
s = '10101'
r = int(s, 2) # первый аргумент - что переводим, второй - из какой системы счисления переводим
print(r)
```

В коде для этой задачи всегда будут три переменные: входное число n, промежуточная двоичная строка s, и результат — десятичное число r.


<h3>Примеры решения</h3>

<b>Пример 1.</b>
<i>"На вход алгоритма подаётся натуральное число N. Алгоритм строит по нему новое число R следующим образом.

Укажите наибольшее число N, после обработки которого с помощью этого алгоритма получается число R, меньшее 100. В ответе запишите это число в десятичной системе счисления."<i>

```python
for n in range(1, 100):
    s = bin(n)[2:]
    if n % 3 == 0:
        s = s + s[-3:]
    else:
        s = s + bin((n % 3) * 3)[2:]
    r = int(s, 2)
    if r < 100:
        print(n, r)
```

В ответе вас могут попросить указать один из двух вариантов: либо число N, либо число R.
<b>Будьте внимательны!</b> В данной задаче просят указать наибольшее число N, так что ответ будет 22.

<b>Пример 2.</b>
<i>"На вход алгоритма подаётся натуральное число N. Алгоритм строит по нему новое число R следующим образом.

Укажите минимальное число N, после обработки которого с помощью этого алгоритма получается число R, большее 40. В ответе запишите это  число в десятичной системе счисления. "<i>

Для того, чтобы <b>заменить</b> какую-то конкретную часть строки (в данном случае два левых разряда), мы сначала <b>срежем их</b>, а затем на их место поместим требуемые.

```python
for n in range(1, 50):
    s = bin(n)[2:]
    if s.count('1') % 2 == 0: #если сумма цифр чётна
        s = '10' + s[2:] + '0'
    else:
        s = '11' + s[2:] + '1'
    r = int(s, 2)
    if r > 40:
        print(n, r)
```

В данной задаче просят указать наибольшее число N, так что ответ будет 16.
