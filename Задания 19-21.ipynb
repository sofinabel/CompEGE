В этом решении играет роль чётность хода. 

В каждой игре <b>первый ход делает Петя</b>. Значит ходы Пети будут соответсвовать порядковым номерам: 1, 3, 5, 7...

Ваня всегда ходит <b>после Пети</b>. Значит ходы Вани будут соответсnвовать порядковым номерам: 2, 4, 6, 8...

В большинстве задач вопросы касаются первых четырёх ходов: победа Пети/Вани первым или вторым ходом.

Если в вопросе задачи сказано, что "Ваня победил после <b>неудачного</b> хода Пети", то <b>all</b> в исходном коде нужно поменять на <b>any</b>.

```python
'''s - текущее кол-во камней в куче curr - сколько ходов сделано сейчас
end - за сколько ходов пытаемся выиграть'''
def f(s, curr, end):
    if s >= 68: #если количество камней в куче выигрышное, то проверяем, наш ли игрок победил
        return curr % 2 == end % 2
    if curr == end: #если текущее кол-во ходов равно максимальному, за которое хотели выиграть, но так и не выиграли
        return 0 #значит и не выиграем

    h = [f(s + 1, curr + 1, end), f(s + 4, curr + 1, end), f(s * 5, curr + 1, end)] #массив следующих ходов
    if (curr + 1) % 2 == end % 2: #если следующий ход наш, то нам достаточно победы хотя бы в одном случае
        return any(h)
    else: #если следующий ход противника, то ему необходимо побеждать во всех ситуациях
        return any(h)


for s in range(1, 68): #перебираю начальное число камней в куче
    for i in range(1, 5): #перебираю число ходов за которое пытаюсь выиграть
        if f(s, 0, i): #если функция вернула истину, значит выигрыш возможен
            print(s, i) #печатаем из какого кол-ва камней и за сколько ходов мы победили
            break #дальше не пытаемся выиграть, так как уже выиграли
```
