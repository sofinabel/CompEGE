Для решения этой задачи будем использовать две функции из библиотеки <b>itertools</b>: для комбинаций будем использовать функцию <b>product</b>, а для перестановок функцию <b>permutations</b>.

<h3>Функция product</h3>

Функция product записывается как:

<b>product(из каких букв выбираем, repeat=длина слова)</b>

```python
for i in product('ABC', repeat=2):
    print(i)
```

Как мы видим, сейчас в цикле мы перебираем просто списки букв, а не слова. Попробуем это исправить.

<h3>Метод join</h3>

Метод join работает с массивами и объединяет символы в одну строку через заданный разделитель. В данной задаче в качестве разделителя используем пустую строку ('').

```python
for i in product('ABC', repeat=2):
    s = ''.join(i)
    print(s)
```

<h3>Примеры решения</h3>

```python
from itertools import *

n = 0 #счётчик для нумерации слов
for i in product('АВОРТ', repeat=6):
    s = ''.join(i) #делаем комбинацию строкой
    n += 1 #увеличиваем порядковый номер
    if s == 'ВОРОТА': #если дошли до требуемого слова
        print(n) #печатаем его номер
```
