<h2>Делы</h2>

Обозначим через ДЕЛ(n, m) утверждение «натуральное число n делится без остатка на натуральное число m». Для какого наибольшего натурального числа А логическое выражение

¬ДЕЛ(x, А) → (ДЕЛ(x, 28) → ¬ДЕЛ(x, 49))

истинно (т.е. принимает значение 1) при любом натуральном значении переменной х?

```python
def f(x, a):
    return (x % a != 0) <= ((x % 28 == 0) <= (x % 49 != 0))

for a in range(1, 1000):
    if all(f(x, a) for x in range(10000)):
        print(a)
```

<h2>Побитовая конъюнкция</h2>

```python
def f(x, a):
    return (x & a == 0) or not(x & 37 == 0) or not (x & 12 == 0)

for a in range(1, 100):
    if all(f(x, a) for x in range(1000)):
        print(a)
```

<h2>Отрезки</h2>

```python
for x in [0.25 * k for k in range(-1000, 1000)]:
    D = 7 <= x <= 68
    C = 29 <= x <= 100
    # если ищем наименьшую длину отрезка, то A = 0
    # если ищем наибольшую длину отрезка, то A = 1
    A = 0
    f = D <= (((not C) and (not A)) <= (not D))
    if f != 1: #ЕСЛИ ВЫРАЖЕНИЕ ДОЛЖНО БЫТЬ ИСТИННО, ТО f != ПРОТИВОПОЛОЖНОМУ А ЗНАЧЕНИЮ, ЕСЛИ ЛОЖНО, ТО f != A
        print(x)
```

<h2>Множества</h2>

```python
for x in [0.25 * k for k in range(-10000, 10000)]:
    P = x in [i for i in range(2, 21, 2)]
    Q = x in [i for i in range(3, 31, 3)]
    # если ищем наименьшую мощность множества, то A = 0
    # если ищем наибольшую мощность множества, то A = 1
    A = 1
    f = (A <= P) and (not(Q) <= (not(A)))
    if f != 0: #если значение функции истинно, то просто меняем цифру A на противоположную
        print(x)
```
