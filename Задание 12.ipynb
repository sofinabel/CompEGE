В этом задании исполнитель Редактор работает со строкой. По условию, у Редактора есть всего две команды:

<i><b>• заменить(v, w).</b> 
Эта команда заменяет в строке первое слева вхождение цепочки v на цепочку w
Если в строке нет вхождений цепочки v, то выполнение команды <b>заменить(v, w)</b> не меняет эту строку</i>

Такая команда есть и в Python. Метод replace работает со строкой и имеет три аргумента: искомая подстрока, замена для подстроки а также число замен. В общем виде метод записывается так:

<h5>строка.replace(что заменяем, на что заменяем, сколько раз)</h5>

```python
s = 'AAA123AAA'
print(s.replace('AAA', '*'))
```

Если не указать последний аргумент, как в примере выше, то будут выполнены замены сразу всех вхождений искомой подстроки в строку. <b>В этом задании последним аргументом всегда пишем 1, то есть заменяем ОДИН раз!</b>

```python
s = 'AAA123AAA'
print(s.replace('AAA', '*', 1))
```

<i><b>• нашлось (v).</b> 
Эта команда проверяет, встречается ли цепочка v в строке исполнителя Редактор. Если она встречается, то команда возвращает логическое значение «истина», в противном случае возвращает значение «ложь». Строка исполнителя при этом не изменяется.</i>

В Python аналогом этой команды будет служить логический оператор <b>in</b>. Запишется это так:
<h5>подстрока in строка</h5>

```python
s = '1234567890'
if '123' in s:
    print('Нашлось!')
```

<h3>Примеры решения</h3>

<b>Пример 1</b>

<i>"Дана программа для редактора:</i>


<i>На вход приведённой выше программе поступает строка, начинающаяся с цифры "2", а затем содержащая n цифр "5" (n > 3).
Определите наименьшее значение n, при котором в строке, получившейся в результате выполнения программы, количество цифр «3» равно 2."</i>

Обратите внимание, что если заменяемые подстроки <b>не пересекаются (никакая из них не является частью другой)</b>, то можно упростить себе жизнь и не писать условия, выделенные жёлтым. В нашем случае это работает: из подстрок "25", "355" и "555" никакая не входит в другую.


```python
for n in range(4, 100):
    s = '2' + n * '5'
    while '25' in s or '355' in s or '555' in s:
        s = s.replace('25', '5', 1)
        s = s.replace('355', '52', 1)
        s = s.replace('555', '3', 1)
    if s.count('3') == 2:
        print(n)
```

Так как по условию требуется найти наименьшее значение n, то ответом в данном случае будет число 18.

<b>Пример 2</b>

<i>"Дана программа для редактора:</i>


<i>На вход приведённой выше программе поступает строка, начинающаяся с цифры 3, а затем n подряд идущих цифр 5 (n > 3).
Определите наименьшее значение n, при котором сумма числовых значений цифр строки, получившейся в результате выполнения программы, равна 17."</i>

В данном задании очень похожий на предыдущий пример алгоритм, однако вопрос чуть более сложный. Необходимо найти <b>сумму цифр строки</b>. Разберём это на более простом примере:

```python
s = '0123456789'
a = 0 #переменная для хранения суммы цифр строки
for i in s: #перебираем символы в строке
    a += int(i) #прибавляем символ к сумме, предварительно сделав его целым числом
print(a)
```

Есть более короткий способ разбить число на цифры с помощью генератора массивов:

```python
s = '0123456789'
a = sum([int(i) for i in s]) #перебираем символы в строке, делаем их целым числом, а затем записываем в массив
print(a)
```

Теперь решим задачу:

```python
for n in range(4, 100):
    s = '3' + n * '5'
    while '25' in s or '355' in s or '555' in s:
        s = s.replace('25', '32', 1)
        s = s.replace('355', '25', 1)
        s = s.replace('555', '3', 1)
    a = sum([int(i) for i in s]) #после работы алгоритма, выйдя из цикла while находим сумму цифр
    if a == 17: #если сумма цифр равна 17, то печатаем n
        print(n)
```

Так как по условию требуется найти наименьшее значение n, то ответом в данном случае будет число 26.
