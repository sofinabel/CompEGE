<b>Всё, что нам нужно заменяем на '*'. Всё, что нам не нужно, заменяем на пробелы, чтобы затем по ним разбить.</b>

```python
f = open('24_8510.txt')
s = f.readline() #считываю весь файл в строку
#всё, что нам нужно, заменяем на *, всё, что нам НЕ нужно, заменяем на пробелы
s = s.replace('N', '*').replace('O', '*').replace('P', '*') #интересующие символы заменяю на пробел
while '**' in s: #пока есть запрещённая комбинация
    s = s.replace('**', '* *') #разбиваю запрещённую комбинацию
s = s.split() #разделяю строку на массив подстрок по пробелам
print(max(s, key=len), len(max(s, key=len)))
```

```python
f = open('24_7094.txt')
s = f.readline()
#заменяю все согл на #
s = s.replace('C', '#').replace('D', '#').replace('F', '#')
#заменяю все гласн на *
s = s.replace('A', '*').replace('U', '*')
#меня интересуют пары согл+гласн то есть пары #*, заменяю их на @
s = s.replace('#*', '@')
#избавляюсь от лишних символов
s = s.replace('#', ' ').replace('*', ' ')
#разбиваю по пробелам
s = s.split()
print(max(s, key=len), len(max(s, key=len)))
```
