В сущности, в этом задании необходимо просто переписать условие на Python и распечатать результат вызова функции

```python
def f(n):
    if n == 1:
        return 1
    if n > 1:
        return f(n - 1) * n
print(f(5))
```

```python
def f(n):
    if n == 1:
        return 1
    if n > 1:
        return n * f(n - 1)
print((f(2023) - f(2022))/f(2020))
```

Здесь иногда возникает ошибка, связанная с переполнением рекурсии. Грубо говоря, число слишком большое, чтобы Python "держал его в уме". Эту ошибку легко исправить Поможем программе, повысив глубину рекурсии.

```python
from sys import *

setrecursionlimit(3000)
def f(n):
    if n == 1:
        return 1
    if n > 1:
        return n * f(n - 1)
print((f(2023) - f(2022))/f(2020))
```
